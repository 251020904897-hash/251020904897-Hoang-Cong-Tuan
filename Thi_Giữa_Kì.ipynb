# **Bài 1 Điều kiện + vòng lặp**
Yêu cầu:

Nhập vào số nguyên dương n.

Tính tổng các số chẵn nhỏ hơn hoặc bằng n.

```python
n = int(input("Nhập số dương n: "))
tong = 0
for i in range(2, n + 1, 2):
  tong += i
print(f"Các số chẵn nhỏ hơn hoặc bằng {n} là: {tong}")
```

# **Bài 2: Chuỗi + xử lý dữ liệu**
Yêu cầu

Nhập vào một chuỗi s.

Thực hiện:

Đếm số từ trong chuỗi

Tìm từ dài nhất

Ví dụ: Input: lap trinh python co ban
       Output: Words: 5 Longest word: python

```python
chu = input("Nhập chuỗi: ")
tach = chu.split()
so_tu = len(tach)
tu_dai_nhat = max(tach, key=len)
print(f"Words: {so_tu}")
print(f"Longest tach: {tu_dai_nhat}")
```

# **Bài 3 List + logic**
Yêu cầu:
 ***Nhập danh sách các số nguyên.***

Yêu cầu:

 ***1. Loại bỏ các phần tử trùng lặp***

 ***2. Sắp xếp danh sách tăng dần***

 ***3. Tìm số lớn thứ hai***

Ví dụ

Input: 5 1 3 5 2 3

Output: Sorted list: [1, 2, 3, 5]

Second largest: 3

```python
so = (input("Nhập số lượng phần tử: "))
danh_sach = []
cac_so = so.split()
for i in cac_so:
  danh_sach.append(int(i))
print("Danh sách: ", danh_sach)
so_khong_trung = []
for i in danh_sach:
  if i not in so_khong_trung:
    so_khong_trung.append(i)
so_khong_trung.sort()
print("Danh sách không trùng lặp: ", so_khong_trung)
if len(so_khong_trung) >= 2:
    so_lon_thu_hai = so_khong_trung[-2]
    print("Second largest:", so_lon_thu_hai)
elif len(so_khong_trung) == 1:
    print("Sorted list:", so_khong_trung)
else:
    print("Sorted list:", so_khong_trung)
```

# **Bài 4 **
Yêu cầu:
***Nhập vào một danh sách điểm sinh viên (0–10).***

Thực hiện:

***Loại bỏ các điểm không hợp lệ***

***Tính điểm trung bình***

***Đếm số sinh viên đạt (≥ 5)***

```python
print("Nhập các điểm sinh viên (cách nhau bởi khoảng trắng):")
chuoi_diem = input()
cac_phan = chuoi_diem.split()
danh_sach_diem = []
for phan in cac_phan:
    so = float(phan)
    danh_sach_diem.append(so)

print("Điểm ban đầu:", danh_sach_diem)
diem_tot = []
for diem in danh_sach_diem:
    if diem >= 0 and diem <= 10:
        diem_tot.append(diem)
print("Điểm hợp lệ:", diem_tot)
tong = 0
for diem in diem_tot:
    tong = tong + diem
so_luong = len(diem_tot)
trung_binh = tong / so_luong
print("Điểm trung bình:", trung_binh)
dem = 0
for diem in diem_tot:
    if diem >= 5:
        dem = dem + 1
print("Đạt:", dem)
```
